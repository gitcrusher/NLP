```python
import numpy as np 
import pandas as pd
import gensim
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
from nltk.corpus import stopwords
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA
import plotly.express as px
import plotly.graph_objects as go
```

```python
story = []
stop_words = set(stopwords.words('english'))
for  filenames in os.listdir('Games_of_throne_books_dataset'): #operates ont the folder.

    # we have used with as an autoclosing statement which means that after the block of code is executed the file will be automatically closed.
    with open(os.path.join('Games_of_throne_books_dataset', filenames)) as f: #  it make "Games_of_throne_books_dataset/book1.txt" and opens in read only mode.
        
        corpus = f.read() # this is gonna read the whole document and store it in corpus variable.
        raw_sentences = sent_tokenize(corpus) # sent_tokenize is gonna split the corpus into sentences and store it in raw_sentences variable.
        for sentences in raw_sentences: 
            words = simple_preprocess(sentences)# simple_preprocess is gonna split the sentences into words, lowercase, removes punctuations and store it in story variable.
            words = [word for word in words if word not in stop_words]
            if len(words)>0:
                story.append(words)
```

```python
model = gensim.models.Word2Vec(
    window = 10,
    min_count  = 2 # ignore words that appear less than 2 times
)
```

```python
model.build_vocab(story) #here we are trying to build vocabulary of the model 
# or in simple words we are trying to tell the model that these are the words which are there in the story and you have to learn about them.
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs) # trianing of model 
```

```python
model.wv.most_similar('daenerys') # here the wv signifies word vector and most_similar is gonna give you the most similar words to daenerys.
# wv is an inbuilt attribute or object of gensim.
```
