```python
import numpy as np 
import pandas as pd
import gensim
import os
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story = []
for  filenames in os.listdir('Games_of_throne_books_dataset'): #operates ont the folder.

    f = open(os.path.join('Games_of_throne_books_dataset', filenames)) #  it make "Games_of_throne_books_dataset/book1.txt" and opens in read only mode.
    corpus = f.read() # this is gonna read the whole document and store it in corpus variable.
    raw_sentences = sent_tokenize(corpus) # sent_tokenize is gonna split the corpus into sentences and store it in raw_sentences variable.
    for sentences in raw_sentences: 
        story.append(simple_preprocess(sentences)) # simple_preprocess is gonna split the sentences into words, lowercase, removes punctuations and store it in story variable.
```
