```python
corpus = []
for i in range(len(sentences)):
    review = re.sub('[^a-zA-Z]',' ',sentences[i])
    review = review.lower()
    corpus.append(review)
```

### Stemming in this corpus

in this technique we are converting the word into its root word. For example, the word "running" will be converted to "run". Stemming is a crude technique and it does not consider the context of the word. Hence it may convert a word into a meaningless root word. for example driking = drik but histoy = histori. Hence stemming is not a very accurate technique.

```python
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word))
```

### Lemmatization

this technique is specially used for converting a word into its root word. It is similar to stemming but it considers the context of the word and converts it into its meaningful root word. For example, the word "better" will be converted to "good" in lemmatization but in stemming it will be converted to "better". Hence lemmatization is more accurate than stemming.

```python
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(Lemmatizer.lemmatize(word))
```

### Implementation of Bag Of Words

here we can see that the corpus is being converted into a matrix of 0's and 1's. This is called as Bag Of Words representation. Here we are considering only the presence of a word in the document and not the frequency of the word. Hence we are using binary=True in CountVectorizer. We can also consider the frequency of the word by setting binary=False.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=True)
```

```python
x = cv.fit_transform(corpus)
```

### N_grams implementation

here we can see that combinations of 3 words are being considered as features. This is called as tri-grams. We can also consider bi-grams and uni-grams.

```python
ng = CountVectorizer(binary=True, ngram_range=(3,3))
```

```python
x = ng.fit_transform(corpus)
```
